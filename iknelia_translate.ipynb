```python
def clean_text(text):
    pass

def translate(text, repeat = False):
    translation = ''
    for word in text.split():
        word = word.lower()
        word = [c for c in list(word) if c.isalpha()]
        word = ''.join(str(x) for x in word)

        options = vocab.get(word, word).split(', ')

        if repeat:
            if len(options) > 1:
                translation += '('
                for i in range(0, len(options)):
                    translation += options[i] + ', '
                translation += ') '
        else:
            translation += options[0] + ' '

    return translation
```

```python
translate('sesenta y nueve', repeat = False)
```

```python
def speech_to_nahuatl(path = 'file.wav'):
    file = speech_rec.AudioFile(path)

    with file as source:
        audio = recognizer.record(source)

    speech = recognizer.recognize_google(audio, language='es-MX')
    trans = translate(speech)

    return speech, trans
```
